```python
# Imports
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

# Load data
dataset = pd.read_csv("/Users/Valina/Documents/DSML-Projects/MLFoundations_CaseStudyApproach/data/ex2data1.txt", names=["x1","x2","y"])

# Prepare input data for scatter plot with Matplotlib

x1_passed = dataset["x1"][dataset["y"]==1]
x2_passed = dataset["x2"][dataset["y"]==1]
x1_notpassed = dataset["x1"][dataset["y"]==0]
x2_notpassed = dataset["x2"][dataset["y"]==0]

# Plot data with Matplotlib

plt.scatter(x1_passed, x2_passed, marker="+", color="g", label="passed")
plt.scatter(x1_notpassed, x2_notpassed, marker="o", color="r", label="not passed")
plt.legend(loc="upper right")
plt.xlabel('Score first exam')
plt.ylabel('Score second exam')
plt.show()
```
